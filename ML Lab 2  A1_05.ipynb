Problem Statement - Housing price prediction with two attributes:Size and Price

```python
# Defining new training data for house sizes (in 1000s of square feet) and corresponding house prices (in 1000s of dollars)
x_train = np.array([1.5, 2.3, 3.8])  # House sizes
y_train = np.array([350.0, 490.0, 765.0])  # House prices

# Displaying the new training data in a more readable format
print("Training Data (House Size -> Price):")
for size, price in zip(x_train, y_train):
    print(f"House Size: {size} (1000 sq.ft) -> Price: ${price}K")
```

```python
## Model function for linear regression
## fw,b(x ) = w + b
```

```python
w = 200
b = 150
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
    x (ndarray (m,)): Data, m examples
    w,b (scalar) : model parameters
    Returns
    f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b,)
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')# Plot our model prediction
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values') # Plot the data points
plt.title("Housing Prices") # Set the title
plt.ylabel('Price (in 1000s of dollars)') # Set the y-axis label
plt.xlabel('Size (1000 sqft)') # Set the x-axis label
plt.legend()
plt.show()
```
